##### 1.1 Write a Python Program(with class concepts) to find the area of the triangle using the below formula.
##### area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
##### Function to take the length of the sides of triangle from user should be defined in the parent class and function to calculate the area should be defined in subclass.

```python
class Shape:
    def area():
        pass
    
class Triangle(Shape):
    def __init__(self,a,b,c):
        self.a = a
        self.b = b
        self.c = c
    def semi_perimeter(self,a,b,c):
        return (a+b+c)/2
    def area(self,a,b,c):
        s = self.semi_perimeter(a,b,c)
        return (s*(s-a)*(s-b)*(s-c)) ** 0.5

a,b,c = 3,4,5
triangle = Triangle(a,b,c)
triangle_area = triangle.area(a,b,c)
print("Area of triangle with sides {}, {}, {} is {}".format(a, b, c, triangle_area))
```

##### 1.2 Write a function filter_long_words() that takes a list of words and an integer n and returns the list of words that are longer than n.

```python
def filter_long_words(list_of_words, n):
    list_of_words_n = []
    for i in list_of_words:
        if(len(i) > n):
            list_of_words_n.append(i)
    return list_of_words_n

list_of_words = input("Enter a sentence :").split()
n = int(input("Enter an integer :"))
filter_long_words(list_of_words,n)
```

##### Here 2,3 and 4 are the lengths of the words in the list.

```python
list_of_words = input("Enter a sentence :").split()

print(list_of_words)
def corresponding_words_lengths(word):
    return len(word)
list(map(corresponding_words_lengths,list_of_words))
########     or     ########
list(map(lambda word: len(word),list_of_words))
```

##### 2.2 Write a Python function which takes a character (i.e. a string of length 1) and returns True if it is a vowel, False otherwise.

```python
character = input("Enter a character :")

def is_vowel(character):
    if(character is 'a' or character is 'e' or character is 'i' or character is 'o' or character is 'u'):
      return True
    return False

print("{} is a vowel? : {}".format(character, is_vowel(character)))
```
