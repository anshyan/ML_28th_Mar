##### 1.1 Write a Python Program to implement your own myreduce() function which works exactly like Python's built-in function reduce()

```python
def myreduce(func, lst):
    result = lst[0]
    for a in lst[1:]:
        result = func(result, a)
    return result
```

```python
myreduce(lambda x,y : x*y, list(range(1,5)))
```

##### 1.2 Write a Python program to implement your own myfilter() function which works exactly like Python's built-in function filter()

```python
def myfilter(func, lst):
    resulted_list = []
    for a in lst:
        if(func(a)):
            resulted_list.append(a)
    return resulted_list
```

```python
myfilter(lambda x: x%2==0, list(range(1,5)))
```

##### ['x', 'xx', 'xxx', 'xxxx', 'y', 'yy', 'yyy', 'yyyy', 'z', 'zz', 'zzz', 'zzzz']

```python
lst = ['x','y','z']
[i*j for i in lst for j in range(1,5)]
```

##### ['x', 'y', 'z', 'xx', 'yy', 'zz', 'xxx', 'yyy', 'zzz', 'xxxx', 'yyyy', 'zzzz']

```python
lst = ['x','y','z']
[i*j for i in range(1,5) for j in lst]
```

##### [[2], [3], [4], [3], [4], [5], [4], [5], [6]] 

```python
lst = [2,3,4,5,6]
[[i+j] for i in lst for j in range(0,3)]
```

##### [[2, 3, 4, 5], [3, 4, 5, 6], [4, 5, 6, 7], [5, 6, 7, 8]]

```python
lst = [2,3,4,5]
[[lst[i]+j for i in range(0,4)] for j in range(0,4)]
```
